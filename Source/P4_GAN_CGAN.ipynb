<h3><u>Información de la práctica</u></h3>
<p style="text-align: justify;">
    Las <strong>redes generativas adversarias (GANs)</strong> nos permiten generar datos como imágenes, vídeos o audio a partir de una entrada aleatoria. Normalmente, esta entrada se muestrea de una distribución normal antes de pasar por una serie de transformaciones que la convierten en algo plausible, como una imagen, un sonido o un clip de vídeo.
  </p>

  <p style="text-align: justify;">
    Sin embargo, una <em>DCGAN</em> sencilla no nos permite controlar la apariencia (por ejemplo, la clase) de las muestras generadas. Por ejemplo, si una GAN genera dígitos manuscritos, no podríamos elegir qué número queremos que aparezca: los resultados serían aleatorios.
  </p>

  <p style="text-align: justify;">
    Para poder controlar lo que se genera, necesitamos <strong>condicionar la salida</strong> de la GAN a una entrada semántica, como la clase de una imagen.
  </p>

  <p style="text-align: justify;">
    En esta práctica (PEC), construiremos una <strong>GAN Condicional</strong> que sea capaz de generar dígitos manuscritos condicionados a una clase concreta.
  </p>

## 3. Cálculo del número de canales de entrada para el generador y el discriminador

En una GAN estándar (no condicional), el generador recibe como entrada un vector de ruido muestreado de una distribución normal, con una dimensión fija (`latent_dim`).

En esta práctica, trabajaremos con una GAN condicional. Eso significa que, además del ruido, también incluiremos información sobre la clase que queremos generar o evaluar. Esta información debe estar presente tanto en el generador como en el discriminador.

Antes de calcular cuántos canales de entrada tendrá cada uno, reflexiona sobre cómo se puede incorporar esta información de clase y qué efecto tiene en el entrenamiento.



```python
# Calcular el número de canales de entrada del generador
# En este caso, se suma la dimensión del vector latente y el número de clases
generator_in_channels = latent_dim + num_classes

# Calcular el número de canales de entrada del discriminador
# Se suma el número de canales de imagen (1 en MNIST) con el número de clases
discriminator_in_channels = num_channels + num_classes

# Mostrar los valores resultantes
print(generator_in_channels, discriminator_in_channels)
```

## 4. Implementación del Generador y del Discriminador

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Pregunta 4.1 [3 pts.]:</strong><br>
Implementa el modelo de Deep Learning utilizando la librería <code>tensorflow.keras</code>. El objetivo es construir un <strong>discriminador</strong> y un <strong>generador</strong> que formen parte de una CGAN.

El discriminador debe ser capaz de distinguir entre imágenes reales y generadas, mientras que el generador debe crear imágenes lo suficientemente realistas como para confundir al discriminador.

Utiliza capas convolucionales, densas y funciones de activación apropiadas. Asegúrate de que el tamaño de entrada y salida sea coherente con el tamaño de las imágenes (28x28x1) y con el vector de entrada del generador (`latent_dim + num_classes`).

Para que el resto del código funcione correctamente, los modelos deben guardarse en variables llamadas <code>generator</code> y <code>discriminator</code>.


### Instrucciones para construir los modelos:

<ul>
  <li>Utiliza <code>Sequential()</code> de <code>keras</code> para definir ambos modelos paso a paso, de forma clara y ordenada.</li>

  <li><strong>Discriminador</strong> (aproximadamente 5–6 capas):
    <ul>
      <li>La entrada debe tener forma <code>(28, 28, discriminator_in_channels)</code>, es decir, la imagen original en escala de grises más la información de clase añadida como canales.</li>
      <li>Incluye varias capas <code>Conv2D</code> con filtros progresivos para extraer características espaciales relevantes.</li>
      <li>Después de algunas convoluciones, añade <code>BatchNormalization</code> para estabilizar el aprendizaje y <code>LeakyReLU</code> como función de activación.</li>
      <li>Se recomienda añadir al menos una capa <code>Dropout</code> para introducir regularización y evitar sobreajuste.</li>
      <li>Al final, utiliza una capa de <code>pooling global</code> para reducir el mapa a un vector y una capa <code>Dense</code> que devuelva un único valor (real o falso).</li>
    </ul>
  </li>

  <li><strong>Generador</strong> (aproximadamente 6 capas):
    <ul>
      <li>La entrada debe tener forma <code>(generator_in_channels,)</code>, es decir, el vector latente concatenado con la clase codificada.</li>
      <li>Comienza con una capa <code>Dense</code> que convierta ese vector en un volumen inicial (por ejemplo, 7×7×canales).</li>
      <li>Usa <code>Reshape</code> para transformar el vector plano en un mapa 2D con profundidad.</li>
      <li>Aplica varias capas <code>Conv2DTranspose</code> (transpuestas) para ir ampliando el tamaño de la imagen de forma progresiva.</li>
      <li>Tras cada transposición, utiliza <code>BatchNormalization</code> y <code>LeakyReLU</code> para mejorar la estabilidad y la calidad de la imagen.</li>
      <li>Finaliza con una <code>Conv2D</code> con activación <code>sigmoid</code> para producir una imagen en escala de grises con valores en el rango [0, 1].</li>
    </ul>
  </li>

  <li>Recuerda utilizar las variables <code>generator_in_channels</code> y <code>discriminator_in_channels</code> definidas anteriormente para que las dimensiones de entrada sean correctas.</li>
</ul>

</div>

```python
# Crear el discriminador.
# Entrada: imagen de 28x28 y 1 canal, más los mapas de clases concatenados (num_classes).
discriminator = keras.Sequential(
    [
        # Entrada del discriminador con 28x28 píxeles y canales = imagen + clases
        keras.layers.InputLayer((28, 28, discriminator_in_channels)),

        # Primera capa convolucional con activación LeakyReLU
        layers.Conv2D(64, (3, 3), strides=(2, 2), padding="same"),
        layers.LeakyReLU(negative_slope=0.2),

        # Segunda capa convolucional con normalización y activación
        layers.Conv2D(128, (3, 3), strides=(2, 2), padding="same"),
        layers.BatchNormalization(),
        layers.LeakyReLU(negative_slope=0.2),

        # Tercera capa adicional con más filtros y regularización
        layers.Conv2D(256, (3, 3), strides=(2, 2), padding="same"),
        layers.BatchNormalization(),
        layers.LeakyReLU(negative_slope=0.2),
        layers.Dropout(0.2),

        # Pooling global y salida final binaria (real/falso)
        layers.GlobalMaxPooling2D(),
        layers.Dense(1),
    ],
    name="discriminator",
)

# Fuentes> https://keras.io/examples/generative/conditional_gan/
```

```python
# Crear el generador.
# Entrada: vector latente + codificación de clase (concatenados).
generator = keras.Sequential(
    [
        # Entrada del generador: vector latente + clase codificada
        layers.InputLayer((generator_in_channels,)),

        # Primera capa totalmente conectada para mapear a 7x7xC
        layers.Dense(7*7*generator_in_channels),
        layers.LeakyReLU(negative_slope=0.2),

        # Reorganizar a mapa 7x7 con canales
        layers.Reshape((7, 7, generator_in_channels)),

        # Primera capa de transposición convolucional (upsample a 14x14)
        layers.Conv2DTranspose(128,(4,4),strides=(2,2),padding='same'),
        layers.BatchNormalization(),
        layers.LeakyReLU(negative_slope=0.2),

        # Segunda capa de transposición convolucional (upsample a 28x28)
        layers.Conv2DTranspose(128,(4,4),strides=(2,2),padding='same'),
        layers.BatchNormalization(),
        layers.LeakyReLU(negative_slope=0.2),

        # Capa final: 1 canal de salida (imagen), con activación sigmoide
        layers.Conv2D(1, (7, 7), padding="same", activation="sigmoid"),
    ],
    name="generator",
)

# Fuentes> https://keras.io/examples/generative/conditional_gan/
```

## 5. Implementación de la clase ConditionalGAN

En esta sección se define una clase personalizada que extiende keras.Model, y que permite controlar manualmente el proceso de entrenamiento de una GAN condicional.

A diferencia de los modelos estándar, una GAN requiere dos redes entrenadas de forma alternada:
- El **discriminador**, que intenta distinguir entre imágenes reales y generadas.
- El **generador**, que intenta crear imágenes lo suficientemente realistas como para “engañar” al discriminador.

El método train_step está estructurado por bloques numerados (apartados) para entender cada etapa del entrenamiento, desde la preparación de datos hasta la actualización de pesos.

Una GAN condicional añade una capa de complejidad: se entrena con imágenes y también con etiquetas que indican la clase deseada.

Lee atentamente el código y responde las siguientes preguntas:


```python
class ConditionalGAN(keras.Model):
    def __init__(self, discriminator, generator, latent_dim):
        super().__init__()
        # Asignar discriminador y generador
        self.discriminator = discriminator
        self.generator = generator

        # Dimensión del vector de entrada aleatorio (ruido)
        self.latent_dim = latent_dim

        # Generador de semillas para reproducibilidad
        self.seed_generator = keras.random.SeedGenerator(1337)

        # Métricas para seguimiento de las pérdidas del generador y del discriminador
        self.gen_loss_tracker = keras.metrics.Mean(name="generator_loss")
        self.disc_loss_tracker = keras.metrics.Mean(name="discriminator_loss")

    @property
    def metrics(self):
        # Lista de métricas que se reportarán durante el entrenamiento
        return [self.gen_loss_tracker, self.disc_loss_tracker]

    def compile(self, d_optimizer, g_optimizer, loss_fn):
        # Configurar optimizadores y función de pérdida
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.loss_fn = loss_fn

    def train_step(self, data):
        # Desempaquetar los datos reales (imágenes y etiquetas one-hot)
        real_images, one_hot_labels = data

        # apartado 1: Expandir y repetir etiquetas
        image_one_hot_labels = one_hot_labels[:, :, None, None]
        image_one_hot_labels = ops.repeat(
            image_one_hot_labels, repeats=[image_size * image_size]
        )
        image_one_hot_labels = ops.reshape(
            image_one_hot_labels, (-1, image_size, image_size, num_classes)
        )

        # apartado 2: Generar vectores latentes aleatorios y concatenarlos con etiquetas
        batch_size = ops.shape(real_images)[0]
        random_latent_vectors = keras.random.normal(
            shape=(batch_size, self.latent_dim), seed=self.seed_generator
        )
        random_vector_labels = ops.concatenate(
            [random_latent_vectors, one_hot_labels], axis=1
        )

        # apartado 3: Generar imágenes falsas a partir del generador
        generated_images = self.generator(random_vector_labels)

        # apartado 4: Concatenar imágenes falsas con sus etiquetas
        fake_image_and_labels = ops.concatenate(
            [generated_images, image_one_hot_labels], -1
        )

        # apartado 5: Concatenar imágenes reales con sus etiquetas
        real_image_and_labels = ops.concatenate([real_images, image_one_hot_labels], -1)

        # apartado 6: Unir el batch de imágenes reales y falsas
        combined_images = ops.concatenate(
            [fake_image_and_labels, real_image_and_labels], axis=0
        )

        # apartado 7: Crear etiquetas: 1 para falsas, 0 para reales
        labels = ops.concatenate(
            [ops.ones((batch_size, 1)), ops.zeros((batch_size, 1))], axis=0
        )

        # apartado 8: Entrenamiento del discriminador
        with tf.GradientTape() as tape:
            predictions = self.discriminator(combined_images)
            d_loss = self.loss_fn(labels, predictions)
        grads = tape.gradient(d_loss, self.discriminator.trainable_weights)
        self.d_optimizer.apply_gradients(
            zip(grads, self.discriminator.trainable_weights)
        )

        # apartado 9: Generar nuevo batch de ruido + etiquetas
        random_latent_vectors = keras.random.normal(
            shape=(batch_size, self.latent_dim), seed=self.seed_generator
        )
        random_vector_labels = ops.concatenate(
            [random_latent_vectors, one_hot_labels], axis=1
        )

        # apartado 10: Etiquetas falsas
        misleading_labels = ops.zeros((batch_size, 1))

        # apartado 11: Entrenamiento del generador
        with tf.GradientTape() as tape:
            fake_images = self.generator(random_vector_labels)
            fake_image_and_labels = ops.concatenate(
                [fake_images, image_one_hot_labels], -1
            )
            predictions = self.discriminator(fake_image_and_labels)
            g_loss = self.loss_fn(misleading_labels, predictions)
        grads = tape.gradient(g_loss, self.generator.trainable_weights)
        self.g_optimizer.apply_gradients(zip(grads, self.generator.trainable_weights))

        # apartado 12: Actualizar métricas de pérdida
        self.gen_loss_tracker.update_state(g_loss)
        self.disc_loss_tracker.update_state(d_loss)

        # Devolver las métricas
        return {
            "g_loss": self.gen_loss_tracker.result(),
            "d_loss": self.disc_loss_tracker.result(),
        }
```

## 7. Interpolación entre clases utilizando el generador entrenado

En esta actividad vamos a analizar cómo el generador responde a entradas que representan clases intermedias (no estrictamente definidas como un único dígito), lo que nos permite explorar la continuidad del espacio latente y semántico aprendido.

Se realizan dos ejercicios:

- **Interpolación entre dos clases**: se seleccionan dos dígitos (por ejemplo, 1 y 9) y se generan etiquetas que van desde una codificación one-hot de la primera clase hasta la segunda. Estas etiquetas interpoladas se combinan con el mismo vector de ruido para producir imágenes intermedias. El resultado es una transición visual entre ambas clases.

- **Generación controlada por clase**: se genera una imagen para cada dígito (del 0 al 9) con ruido aleatorio diferente, y se observa si el generador ha aprendido a producir ejemplos coherentes para cada clase concreta.

El resultado se visualiza como:
- Un GIF animado para la interpolación entre clases.
- Una figura con 10 imágenes generadas en una sola fila, una por clase.


```python
# Extraemos el generador entrenado de nuestro CGAN
trained_gen = cgan_model.generator  # Cambiado al nuevo nombre del modelo

# Número total de imágenes interpoladas entre dos clases (incluye inicio y fin)
num_interpolation = 9

# Generamos ruido aleatorio para cada punto intermedio de la interpolación
interpolation_noise = keras.random.normal(shape=(1, latent_dim))
interpolation_noise = ops.repeat(interpolation_noise, repeats=num_interpolation)
interpolation_noise = ops.reshape(interpolation_noise, (num_interpolation, latent_dim))
```

```python
# Función que genera imágenes interpoladas entre dos clases
def interpolate_class(first_number, second_number):
    # Convertimos las clases inicial y final a codificación one-hot
    first_label = keras.utils.to_categorical([first_number], num_classes)
    second_label = keras.utils.to_categorical([second_number], num_classes)
    first_label = ops.cast(first_label, "float32")
    second_label = ops.cast(second_label, "float32")

    # Interpolamos entre las dos etiquetas (línea recta en el espacio de clases)
    percent_second_label = ops.linspace(0, 1, num_interpolation)[:, None]
    percent_second_label = ops.cast(percent_second_label, "float32")
    interpolation_labels = (
        first_label * (1 - percent_second_label) + second_label * percent_second_label
    )

    # Concatenamos ruido + etiquetas interpoladas y generamos imágenes
    noise_and_labels = ops.concatenate([interpolation_noise, interpolation_labels], 1)
    fake = trained_gen.predict(noise_and_labels)
    return fake

# Especificamos las clases entre las que queremos interpolar
start_class = 1  # Clase inicial
end_class = 9    # Clase final

# Generamos las imágenes falsas correspondientes a la interpolación
fake_images = interpolate_class(start_class, end_class)
```

Mostramos ruido de una distribución normal y después lo repetimos para num_interpolation veces y damos forma al resultado en consecuencia. A continuación lo distribuimos uniformemente para num_interpolation con las etiquetas presentes proporcionalmente.

```python
# Escalamos las imágenes generadas al rango [0, 255] y convertimos a uint8
fake_images *= 255.0
converted_images = fake_images.astype(np.uint8)

# Redimensionamos las imágenes a 144x144 píxeles
converted_images = ops.image.resize(converted_images, (144, 144)).numpy().astype(np.uint8)

# Guardamos las imágenes en un archivo GIF con una imagen por segundo
imageio.mimsave("animation.gif", converted_images[:, :, :, 0], fps=1)

# Mostramos el GIF directamente en el notebook
embed.embed_file("animation.gif")
```

```python
import matplotlib.pyplot as plt

# Generar ruido aleatorio para cada clase
num_samples = 10  # Por ejemplo, una imagen por cada clase del 0 al 9
random_latent_vectors = keras.random.normal(shape=(num_samples, latent_dim))

# Crear etiquetas one-hot para las clases del 0 al 9
one_hot_labels = keras.utils.to_categorical(np.arange(num_samples), num_classes)

# Concatenar ruido + etiquetas y generar imágenes
noise_and_labels = ops.concatenate([random_latent_vectors, one_hot_labels], axis=1)
generated_images = trained_gen.predict(noise_and_labels)

# Escalar las imágenes al rango [0, 255] para visualización
generated_images *= 255.0
generated_images = generated_images.astype(np.uint8)

# Redimensionar a 144x144 si se desea visualización grande
generated_images_resized = ops.image.resize(generated_images, (144, 144)).numpy().astype(np.uint8)

# Mostrar las imágenes en una sola fila
plt.figure(figsize=(15, 3))
for i in range(num_samples):
    plt.subplot(1, num_samples, i + 1)
    plt.imshow(generated_images_resized[i, :, :, 0], cmap="gray")
    plt.axis("off")
    plt.title(f"Clase {i}")
plt.tight_layout()
plt.show()
```
